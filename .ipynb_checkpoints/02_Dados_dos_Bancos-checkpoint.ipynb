```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
dados = pd.read_csv('C:/Users/vitor/Desktop/Dados_Banco.csv')
dados
```

# Alterando Colunas / Resetando Index / Nomeando Index / Exibindo repetições

```python
bancos = pd.DataFrame(dados['Banco'].value_counts())
bancos.reset_index(inplace = True)
bancos.columns.name ='ID'
bancos.rename(columns = {'Banco':'Quantidade','index':'Bancos'},inplace = True)
bancos
```

```python
n = dados.shape[0]
k = 1 + (10 /3) * np.log10(n) 
k = int(k)
k
```

```python
freq1 = pd.value_counts(pd.cut(x = dados.Renda, bins = k, include_lowest = True), sort = False)
perc1 = pd.value_counts(pd.cut(x = dados.Renda, bins = k, include_lowest = True), sort = False, normalize = True)*100
```

```python
frequencia_de_salarios =  pd.DataFrame({'Frequência':freq1,'Porcentagem (%)':perc1}).rename_axis('Salarios', axis = 'columns')
frequencia_de_salarios.reset_index(inplace = True)
frequencia_de_salarios.columns
frequencia_de_salarios.rename(columns = {'index':'Intervalos'}, inplace = True)
frequencia_de_salarios.columns.name = 'ID'
frequencia_de_salarios
```

```python
dados['Limite_Saque'].fillna(value = 'Não', inplace = True)
dados[(dados.Banco == 'Ouro')]
# E POSSIVEL NOTAR QUE PESSOAS COM MAIOR IDADE E MAIS EXPERIENCIA TEM UMA MAIOR RENDA E ESCOLHEM O BANCO OURO
```

```python
dados[(dados.Banco == 'Bronze')]
# E POSSIVEL NOTAR QUE PESSOAS COM MENOR IDADE E MAIS EXPERIENCIA TEM UMA MENOR RENDA E ESCOLHEM O BANCO BRONZE
```

```python
dados2 =  pd.DataFrame(dados.query('Banco in ["Bronze","Ouro"]'))
```

```python
dados2
```

```python
troca = {'Sim':1 ,'Não':0}

dados2['Poupança'] = dados2['Poupança'].map(troca)
dados2['Emprestimo'] = dados2['Emprestimo'].map(troca)
dados2['Plano_Previdencia'] = dados2['Plano_Previdencia'].map(troca)
dados2['Limite_Saque'] = dados2['Limite_Saque'].map(troca)
dados2
```

```python
x = dados2[['Poupança','Emprestimo','Plano_Previdencia','Limite_Saque']]
y = dados2[['Banco']]

treino_x = x[:18]
treino_y = y[:18]
teste_x = x[18:]
teste_y = x[18:]
```

```python
from sklearn.model_selection import train_test_split
from sklearn.svm import LinearSVC
from sklearn.metrics import accuracy_score

seed = 20

treino_x, teste_x, treino_y, teste_y = train_test_split(x, y,
                                                        random_state = 20,
                                                        test_size = 0.8,
                                                        stratify = y)
```

```python
modelo = LinearSVC()
modelo.fit(treino_x, treino_y.values.ravel())
previsoes = modelo.predict(teste_x)
previsoes
```

```python
acuracia = accuracy_score(teste_y, previsoes) * 100
print("A acurácia foi %.2f%%"  % acuracia)
```
