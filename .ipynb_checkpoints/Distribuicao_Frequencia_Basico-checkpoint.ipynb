```python
# Regra de Sturges otimiza a escolha da quantidade de classes que teremos nas tabelas de distribuições, 
# considerando somente o número de observações que temos na variável.
# # k = 1 + (10 /3) * np.log10(n)  

#Frequencias estao em classes
```

```python
import numpy as np
```

```python
dados.shape
```

```python
n = dados.shape[0]
```

```python
k = 1 + (10 /3) * np.log10(n) 
k = int(k)
k
```

```python
frequencia_01 = pd.value_counts(pd.cut(x = dados.Renda, bins = k, include_lowest = True), sort = False)
porcentual_01 = pd.value_counts(pd.cut(x = dados.Renda, bins = k, include_lowest = True),sort = False, normalize = True)
```

```python
distribuicao_freq_quantitativa = pd.DataFrame({'Frequencia':frequencia_01, 'Porcentual':porcentual_01})
distribuicao_freq_quantitativa.rename_axis('Rendas', axis = 'columns', inplace = True)
distribuicao_freq_quantitativa
```

```python
n = frame.shape[0]
v = 1 + (10 /3) * np.log10(n) 
v = int(v)
v
```

```python
pd.cut(x = frame.Gasto_Mes, bins = v, include_lowest = True) #Vai repitir todas as vezes que intervalos aparecem, e não o numero de interlavos
```

```python
frequencia_frame = pd.value_counts(pd.cut(x = frame.Gasto_Mes, bins = v, include_lowest = True),sort = False)
frequencia_frame
```

```python
porcentual_frame = pd.value_counts(pd.cut(x = frame.Gasto_Mes, bins = v, include_lowest = True),sort = False, normalize = True)*100
porcentual_frame
```

```python
tabela_de_distribuicao_freq = pd.DataFrame({'Frequencia':frequencia_frame,'Porcentual':porcentual_frame})
```

```python
tabela_de_distribuicao_freq
```

```python
tabela_de_distribuicao_freq.sort_index(ascending = False, inplace = True)
```

```python
tabela_de_distribuicao_freq.rename_axis('Intervalo_Classes', axis = 'columns',inplace = True)
tabela_de_distribuicao_freq
```
