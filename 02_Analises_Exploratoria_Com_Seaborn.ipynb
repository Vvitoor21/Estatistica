```python
import pandas as pd
gorjetas = pd.read_csv('C:/Users/vitor/Desktop/Estudos/tips_atualizado.csv')
gorjetas
```

```python
colunas = {'total_bill':'conta','tip':'gorjeta','dessert':'sobremesa','day':'dia','time':'hora','size':'pessoas'}
gorjetas.rename(columns = colunas, inplace = True)
gorjetas
```

```python
gorjetas.dia.unique()
```

```python
sobremesa = {'No':'Não','Yes':'Sim'}
dia = {'Sun':'Domingo','Sat':'Sábado','Thur':'Quinta','Fri':'Sexta'}
hora = {'Dinner':'Jantar','Lunch':'Almoço'}

gorjetas['sobremesa'] = gorjetas['sobremesa'].map(sobremesa)
gorjetas['dia'] = gorjetas['dia'].map(dia)
gorjetas['hora'] = gorjetas['hora'].map(hora)

gorjetas
```

```python
gorjetas['percentual'] = (gorjetas['gorjeta'] / gorjetas['conta']).round(2)
gorjetas
```

```python
from scipy.stats import ranksums
valor_conta_domingo = gorjetas.query("dia == 'Domingo'").conta
valor_conta_sabado = gorjetas.query("dia == 'Sábado'").conta
```

```python
r2 = ranksums(valor_conta_domingo, valor_conta_sabado)
print('O valor do p-value é {}'.format(r2.pvalue)) #P value menor que 0.05 aponta para uma hipotese alternativa.
                                                   #O média do valor da conta é maior no sábado
```
